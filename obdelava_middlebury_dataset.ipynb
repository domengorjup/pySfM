```python
import os
import glob
import time
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm, tqdm_notebook
import tools
```

### komentarji:
    
    - unikaten identifier vsake matrike ``[R|t]`` (da niso naslavljane po indeksu)

```python
SCALE = 0.5
MATCH_RATIO = 0.85 # more -> LESS matches rejected
NFEV = 3000
max_outlier_sd = 1
```

```python
path = 'datasets\\dinoRing'
images = glob.glob(os.path.join(path, '*.png'))
txts = glob.glob(os.path.join(path, '*.txt'))
parameters_file = [_ for _ in txts if '_par' in _][0]
```

### main loop

```python
beginning_time = time.perf_counter()
```

```python
cameras = {}
cameras['K'] = tools.camera_matrices_from_txt(parameters_file)[1]
cameras['N'] = len(images)
cameras['image'] = images
cameras['P'] = tools.camera_matrices_from_txt(parameters_file)[0]
```

```python
plot = True
%matplotlib inline

for i in tqdm_notebook(range(cameras['N'] - 1)):
    Rt1 = cameras['P'][i]
    P1 = cameras['K'].dot(Rt1)
    Rt2 = cameras['P'][i + 1]
    P2 = cameras['K'].dot(Rt2)
    im1 = tools.read_bw(cameras['image'][i])
    im2 = tools.read_bw(cameras['image'][i+1])
    p1, p2 = tools.match_features(im1, im2, scale=SCALE, match_ratio=MATCH_RATIO, mode='ORB')
    
    # reject outliers through RANSAC
    F, mask = cv2.findFundamentalMat(p1, p2, cv2.FM_RANSAC, 1, 0.995)
    p1 = np.array([_ for i, _ in enumerate(p1) if mask[i]])
    p2 = np.array([_ for i, _ in enumerate(p2) if mask[i]])
    
    points3D = []
    for x1, x2 in zip(p1, p2):
        X = tools.triangulate_lsq(tools.to_homogenous(x1), tools.to_homogenous(x2), P1, P2)
        points3D.append(tools.Point3D(X, x1, x2, i, i+1))
    
    if i == 0:
        cloud = tools.Scene([Rt1, Rt2], points3D, cameras['K'], debug=True, shape=im1.shape)
        cloud.bundle_adjustment(ftol=1e-5, max_nfev=NFEV, max_sd_dist=max_outlier_sd, X_only=True)
    else:
        pair = tools.Scene([Rt1, Rt2], points3D, cameras['K'], frame_offset=i, debug=True, shape=im1.shape)
        pair.bundle_adjustment(ftol=1e-5, max_nfev=NFEV, max_sd_dist=max_outlier_sd, X_only=True)
        cloud.add_next_pair(pair)
        
    if plot and i % (cameras['N']//3) == 0:
        cloud.render()
        
        fig, ax = plt.subplots(1, 2, figsize=(15, 3))
        ax[0].imshow(im1, cmap='gray')
        ax[0].scatter(p1[:, 0], p1[:, 1])
        ax[1].imshow(im2, cmap='gray')
        ax[1].scatter(p2[:, 0], p2[:, 1])
        plt.show()
        
    #print('{:5.1f} %'.format((i+1) / (cameras['N']-1) * 100), end='\r')
```

```python
main_loop_time = time.perf_counter()
```
